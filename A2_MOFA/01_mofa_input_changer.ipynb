```python
omics = pd.read_csv("./multiome_last_mofa-data.csv", low_memory=False)
omics = pd.DataFrame(omics)
```

```python
import pandas as pd
import re

# group_1 (unexposed normal group) was deleted. 
omics_comp1 = omics[omics['group'] != 'group_1'].copy()
```

```python
omics_comp1
```

```python
# Change group
omics_comp1.loc[:, 'group'] = omics_comp1['group'].replace({
    'group_2': 'group_1',
    'group_3': 'group_2',
    'group_4': 'group_2',
    'group_5': 'group_3'
})
```

```python
omics_comp1
```

```python
# Change group information of sample column - using regular expression
def update_sample(sample_name):
    match = re.search(r'_group_(\d+)', sample_name)
    if match:
        group_number = int(match.group(1))
        if group_number == 2:
            new_group_number = 1
        elif group_number in [3, 4]:
            new_group_number = 2
        elif group_number == 5:
            new_group_number = 3
        else:
            return sample_name
        return re.sub(r'_group_\d+', f'_group_{new_group_number}', sample_name)
    return sample_name

omics_comp1.loc[:, 'sample'] = omics_comp1['sample'].apply(update_sample)

print(omics_comp1)
```

```python
omics_comp1 = omics_comp1.drop(columns=['Unnamed: 0'])
omics_comp1.index = range(1, len(omics_comp1)+1)

print(omics_comp1)
```

```python
omics_comp1.to_csv("./multiome_last_comp1_data.csv")
```
